Here we extracted a subset of our selected columns. Additionally, we filtered this subset to gain a dataset without missing values.

```python
patients_2=df.loc[:, ['Alcohol', 'Country', 'infant deaths','Polio','Measles', 'Year','Schooling']]
```

```python
patients_2 = df.dropna(thresh=6, subset=['Year', 'Alcohol', 'infant deaths', 'Polio','Measles','Schooling'])
df_subset = patients_2.loc[:, ['Year', 'Alcohol', 'Country', 'infant deaths', 'Polio','Measles','Schooling']]
df_subset
```

This code was used to check whether the missing values were truly removed.

```python
df_subset.isnull().sum()
```

Here we created an overview by orderning per country.

```python
countries_of_interest = ['Netherlands', 'Cameroon', 'Australia', 'Zimbabwe']
filtered_df = df_subset[df_subset['Country'].isin(countries_of_interest)]
filtered_df = filtered_df[['Year', 'Country', 'infant deaths', 'Alcohol', 'Polio','Measles','Schooling']]
print(filtered_df)
```

```python
sorted_df = filtered_df.sort_values(by='Country')
print(sorted_df)
```

A scatterplot was created to obtain insight on the changes in year per variable. To make it more clear we coloured the different countries.

```python
alt.Chart(filtered_df).mark_point().encode(
    x=alt.X(('Year:Q'),scale=alt.Scale(domain=(1999,2016))),
    y=alt.Y('Alcohol:Q'),
    color='Country:N'
).properties(title='Alcohol per year per country')|alt.Chart(filtered_df).mark_point().encode(
    x=alt.X(('Year:Q'),scale=alt.Scale(domain=(1999,2016))),
    y=alt.Y('Polio:Q'),
    color='Country:N'
).properties(title='Polio per year per country')|alt.Chart(filtered_df).mark_point().encode(
    x=alt.X(('Year:Q'),scale=alt.Scale(domain=(1999,2016))),
    y=alt.Y('Measles:Q'),
    color='Country:N'
).properties(title='Measles per year per country')|alt.Chart(filtered_df).mark_point().encode(
    x=alt.X(('Year:Q'),scale=alt.Scale(domain=(1999,2016))),
    y=alt.Y('infant deaths:Q'),
    color='Country:N'
).properties(title='infant deaths per year per country')|alt.Chart(filtered_df).mark_point().encode(
    x=alt.X(('Year:Q'),scale=alt.Scale(domain=(1999,2016))),
    y=alt.Y('Schooling:Q'),
    color='Country:N'
).properties(title='Schooling per year per country')
```

Wanting to understand the scatterplot more, due to overlapping, we created bar charts of max values per country. These helped get a better understanding of the dataset variables, and compare the variables to each other.

```python
alt.Chart(filtered_df).mark_bar().encode(
    x='Country',
    y='max(Alcohol)'
)|alt.Chart(filtered_df).mark_bar().encode(
    x='Country',
    y='max(Polio)'
)|alt.Chart(filtered_df).mark_bar().encode(
    x='Country',
    y='max(Measles)'
)|alt.Chart(filtered_df).mark_bar().encode(
    x='Country',
    y='max(infant deaths)'
)|alt.Chart(filtered_df).mark_bar().encode(
    x='Country',
    y='max(Schooling)'
)
```
